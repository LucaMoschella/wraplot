```python
import numpy as np
import matplotlib
matplotlib.use("Agg")

from wraplot.plotter import *
```

```python
sparse_matrix = (np.random.rand(100, 100) > .5).astype(np.float)
dense_matrix = np.random.rand(100, 100).astype(np.float)
points = np.random.rand(100, 3)
```

# Spy

```python
spyplot = Spy(figsize=(5, 5))
spyplot(Spy.Object(matrix=sparse_matrix, markersize=4))
```

# Imagesc

```python
imagescplot = Imagesc(figsize=(5, 5))
imagescplot(Imagesc.Object(matrix=dense_matrix))
```

# Plotcloud2D

```python
cloudplotter = PlotCloud2D(figsize=(5, 5))
cloudplotter(PlotCloud2D.Object(points=points))
```

# Plot bar comparison

```python
barplotter = PlotCoupledBarsComparison()
o = barplotter.Object(
    bar1_values=['adf', 'sf'],
    bar2_values=[1, 3],
    label1 = 'a',
    label2 = 'b',
    xticklabels=['1a', '1b'],
    
    title='ciao',
    titlesize=50,
    
    subtitle='ooo'
)
# barplotter(o)
sub = Subplotter()

sub([o, o], plot_functions=barplotter, subplotadjust=0.5)
```
